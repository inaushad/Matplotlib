```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos=np.arange(len(company))
```

```python
plt.title('US Technolog Stocks')
plt.ylabel("Revenue(in bin)")
plt.xlabel("Company Name")
plt.xticks(xpos,company)
plt.bar(company,revenue,label="Revenue")
plt.legend()
```

```python
profit=[40,2,34,12]
share=[45,12,26,30]
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue(in bin)")
plt.bar(xpos-0.2,revenue,width=0.2,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.2,label="Profit")
plt.bar(xpos,share,width=0.2,label="Share")
plt.xticks(xpos,company)
plt.legend()
```

```python
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```
