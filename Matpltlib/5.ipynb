```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=[80,100,125,150],color='g',rwidth=0.8)
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women=[67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('Sugar Level')
plt.ylabel('No of Patients')
plt.title('Blood Sugar Chart')
plt
plt.hist([blood_sugar_men,blood_suger_women], bins=[80,100,125,150],color=['green','orange'],rwidth=0.8,label=['Men','Women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_suger_women=[67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
plt.xlabel('Sugar Level')
plt.ylabel('No of Patients')
plt.title('Blood Sugar Chart')
plt
plt.hist([blood_sugar_men,blood_suger_women], bins=[80,100,125,150],color=['green','orange'],rwidth=0.8,label=['Men','Women'],orientation='horizontal')
plt.legend()
```
