```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python

plt.pie(exp_vals,labels=exp_labels,shadow=True)
plt.axis("equal")
plt.show()
```

```python

plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%1.1f%%',radius=1.5)
plt.axis("equal")
plt.show()
```

```python

plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct='%1.1f%%',radius=1.5,explode=[0.5,0.4,0.3,0.1,0.2])
plt.axis("equal")
plt.show()
plt.savefig("pie,jpg",bbox_inches="tight",pad_inches=1, transparent=True)
```
